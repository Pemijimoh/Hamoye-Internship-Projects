### Question 12
From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
# temperature in the living room
x = df['T2']

# temperature outside the building
y = df['T6']
```

```python
# reshape data to 2D array
data = np.array(x)
x = data.reshape(-1, 1)
```

```python
#splitting the data into train and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size = 0.3, random_state=42)

#Define the model
reg_model = LinearRegression()

# fit the model
reg_model.fit(x_train, y_train)
pred = reg_model.predict(x_test)
```

```python
# find r2 score

from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred)
round(r2_score, 2)
```

#### Answer
The r squared score is 0.64.

### Question 13
Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:

```python
# drop 'light' and 'data' variable
df = df.drop(['lights', 'date'], axis = 1)
```

```python
# normalize the dataset using min max scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

features = normalized_df.drop(columns=['Appliances'])
target = normalized_df['Appliances']
```

```python
# split the normalized data into test and train set
xn_train, xn_test, yn_train, yn_test = train_test_split(features, target, train_size=0.7, test_size = 0.3, random_state=42)
```

```python
#Define our model
reg_model2 = LinearRegression()

# fit the model
reg_model2.fit(xn_train, yn_train)

# predict on test set
predicted_value = reg_model2.predict(xn_test)
```

```python
# evaluating the model based on mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(yn_test, predicted_value)
round(mae, 2)
```

#### Answer
The mean absolute error is 0.05.

### Question 14
What is the Residual Sum of Squares (in two decimal places)?

```python
# calculate RSS value
residual = np.sum((yn_test - predicted_value)**2)
round(residual, 2)
```

#### Answer
The residual sum of squares (RSS) is 45.35

### Question 15
What is the Root Mean Squared Error (in three decimal places)?

```python
# find mean squared error

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(yn_test, predicted_value))
round(rmse, 3)
```

#### Answer
The root mean squared error is 0.088.

### Question 16
What is the Coefficient of Determination (in two decimal places)?

```python
# find r squared score

from sklearn.metrics import r2_score
r2_score = r2_score(yn_test, predicted_value)
round(r2_score, 2)
```

#### Answer
The coefficient of determination is 0.15

### Question 17
Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
# obtaining feature weights from linear model

def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
# get weights
linear_model_weights = get_weights_df(reg_model2, xn_train, 'Linear_Model_Weight')
linear_model_weights
```

#### Answer
From the dataframe above, RH_2 has the lowest weight while RH_1 has the highest weight.

### Question 18
Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

### Question 19
Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
# number of features with non-zero weights

non_zero_features = lasso_weights_df.query('Lasso_weight != 0')
len(non_zero_features)
```

#### Answer
